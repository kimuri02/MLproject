```python
import os
import pandas as pd

os.chdir('C:/Users/rameo/Desktop/ajou/ML_project') 
os.getcwd()

# 날씨 데이터 불러오기
weather = pd.read_csv('OBS_ASOS_DD_20220602140342.csv', sep=',', encoding='CP949')
print(weather.info())
weather.head()
```

```python
weather.isnull().sum() 
```

```python
# 사용할 열 추출
# use_col = ['평균기온(°C)','일강수량(mm)','평균 풍속(m/s)','평균 이슬점온도(°C)',
#            '평균 상대습도(%)','평균 증기압(hPa)','평균 지면온도(°C)']

use_col = ['일시','평균기온(°C)','일강수량(mm)','평균 풍속(m/s)','평균 풍속(m/s)','평균 이슬점온도(°C)','평균 상대습도(%)',
         '평균 증기압(hPa)','평균 현지기압(hPa)','평균 해면기압(hPa)','합계 일사량(MJ/m2)','평균 전운량(1/10)',
         '평균 중하층운량(1/10)','평균 지면온도(°C)','합계 대형증발량(mm)','합계 소형증발량(mm)']

weather_use = weather.loc[:,use_col]

# 결측값 확인
weather_use.values.any() 
pre_weather = weather_use.fillna(method='pad') # 앞의 행 값으로 대체
pre_weather.isnull().sum() 

# target 데이터 생성

# 봄 : 3~5월,  여름 : 6~8월, 가을 : 9~11월, 겨울 : 12~2월
spring = ['03', '04', '05']
summer = ['06', '07', '08']
fall = ['09', '10', '11']
winter = ['12', '01','02']

target = []

for i in range(len(weather)):    
    month = weather.loc[:,'일시'][i].split('-')[1]

    if month in spring:
        target.append(0)
    elif month in summer:
        target.append(1)
    elif month in fall:
        target.append(2)
    elif month in winter:
        target.append(3)

weather['y'] = target
pre_weather.iloc[:,1:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from graphviz import Source
from sklearn.tree import export_graphviz


# 데이터 나누기
X_train, X_test, y_train, y_test = train_test_split(pre_weather, weather['y'], test_size=0.3, random_state=11)


# decision tree
decision_model = DecisionTreeClassifier(max_depth=3, random_state=11)
decision_model.fit(X_train.iloc[:,1:], y_train)

pred_train = decision_model.predict(X_train.iloc[:,1:])
pred_test = decision_model.predict(X_test.iloc[:,1:])
# print(pred_train)
# print(pred_test)

# visualize
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

label = ['spring', 'summer', 'fall', 'winter']
export_graphviz(
        decision_model,
        out_file=os.path.join(IMAGES_PATH, "decision_model.dot"),
        feature_names=use_col[1:],
        class_names=label,
        rounded=True,
        filled=True
    )
Source.from_file(os.path.join(IMAGES_PATH, "decision_model.dot"))
```

```python
print(X_test.iloc[:,1:])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# accuracy
print('train accuracy : ', accuracy_score(pred_train, y_train))
print('test accuracy : ', accuracy_score(pred_test, y_test))

# confusion matrix
confusion_matrix(y_test,pred_test)
```

# 사고 유형

```python
accident_type_raw = pd.read_excel('accident type.xls')

accident_type = accident_type_raw.drop(accident_type_raw.index[366])

accident_type.isnull().sum()
# accident_type
```

```python
accident_type_raw
```

```python
date = []

for i in range(len(accident_type)):
    m = accident_type['사고월'][i].split('월')[0]
    d = accident_type['사고일'][i].split('일')[0]
    
    date.append('2020-' + m + '-' + d)

accident_type = accident_type.drop(['사고월','사고일'], axis='columns')
accident_type['일시'] = date
accident_type = accident_type[['일시', '횡단중','정면충돌','추돌','주/정차차량 충돌','도로이탈','전복']]
accident_type = accident_type.set_index('일시')
accident_type
```

```python
spring_df = pd.DataFrame()
summer_df = pd.DataFrame()
fall_df = pd.DataFrame()
winter_df = pd.DataFrame()

# 봄 교통사고 유형
for i in range(len(spring)):
    ds = accident_type[accident_type.index == spring[i]]
    spring_df = pd.concat([spring_df,ds])
# spring_df

# 여름 교통사고 유형
for i in range(len(summer)):
    ds = accident_type[accident_type.index == summer[i]]
    summer_df = pd.concat([summer_df,ds])
# summer_df

# 가을 교통사고 유형
for i in range(len(fall)):
    ds = accident_type[accident_type.index == fall[i]]
    fall_df = pd.concat([fall_df,ds])
# fall_df

# 겨울 교통사고 유형
for i in range(len(winter)):
    ds = accident_type[accident_type.index == winter[i]]
    winter_df = pd.concat([winter_df,ds])
# winter_df

print('-------------------------\n')
print('spring\n')
print(spring_df.mean(),'\n')
print('-------------------------\n')
print('summer\n')
print(summer_df.mean(),'\n')
print('-------------------------\n')
print('fall\n')
print(fall_df.mean(),'\n')
print('-------------------------\n')
print('winter\n')
print(winter_df.mean(),'\n')
print('-------------------------\n')
```
